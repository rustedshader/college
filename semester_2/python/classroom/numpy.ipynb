```python
import pandas as pd
import matplotlib.pyplot as plt
np.random.seed(69)
customers_id = np.random.randint(low=0,high=100,size=10)


balance = {int(_):np.random.randint(low=0,high=10000,size=30) for _ in customers_id}
df = pd.DataFrame(balance)
for _ in balance:
    df.loc['avg'] = df.mean()
print(df)
plt.plot([_ for _ in balance],[balance[_] for _ in balance])
plt.show()
plt.bar([_ for _ in customers_id] , df.loc['avg'])
plt.show()
```
